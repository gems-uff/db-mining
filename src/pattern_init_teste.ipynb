```python
#Renaming collumn 
#df13.columns.rename("unnamed: 0","")
#df13.drop("unnamed: 0", axis=0, inplace=True)
df13.rename(columns = {'Unnamed: 0':'PID'}, inplace = True)
df13
```

Calculateding confidence and lift to each pattern generated.

```python
#Converting to integer
df13['SUP_A'] = df13['SUP_A'].astype(int)
df13['SUP_B'] = df13['SUP_B'].astype(int)
df13['SUP_C'] = df13['SUP_C'].astype(int)
df13['SUP_D'] = df13['SUP_D'].astype(int)
df13['SUP_BC'] = df13['SUP_BC'].astype(int)
df13['SUP_BCD'] = df13['SUP_BCD'].astype(int)
```

```python
#df13.drop("ConfA->_BCD", axis=1, inplace=True)
#df13.drop("ConfB->A_CD", axis=1, inplace=True)
#df13.drop("ConfC->AB_D", axis=1, inplace=True)
#df13.drop("ConfD->ABC_", axis=1, inplace=True)
```

```python
#Inserting collums to confidence measure
df13.insert(8,'ConfA->_BCD','',False) 
df13.insert(11,'ConfB->A_CD','',False)
df13.insert(14,'ConfC->AB_D','',False)
df13.insert(17,'ConfD->ABC_','',False)
df13
```

Calculeting the confidence measures

```python
df13['ConfA->_BCD'] = (df13['SUP_ABCD'])/df13['SUP_A']
df13['ConfB->A_CD'] = (df13['SUP_ABCD'])/df13['SUP_B']
df13['ConfC->AB_D'] = (df13['SUP_ABCD'])/df13['SUP_C']
df13['ConfD->ABC_'] = (df13['SUP_ABCD'])/df13['SUP_D']
df13
```

Calculeting the lift measures

```python
df13['LiftA->B'] = (df13['SUP_ABCD']/200)/((df13['SUP_A']/200)*(df13['SUP_B']/200))
df13['LiftA->BC'] = (df13['SUP_ABCD']/200)/((df13['SUP_A']/200)*(df13['SUP_BC']/200))
df13['LiftA->BCD'] = (df13['SUP_ABCD']/200)/((df13['SUP_A']/200)*(df13['SUP_BCD']/200))
df13
```

```python
#Generating excel with selected INIT patterns and interest measures
df13.to_excel(SEQ_PATTERNS_DIR + os.sep +"pattern_supABCD_init_measure.xlsx",index=True) 
```
