```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from util import COUNT_FILE_SQL
#reads projects from Excel file
df = pd.read_excel(COUNT_FILE_SQL, keep_default_na=False)
df = df.rename(columns={"(IgnoreCase, SQL)": "SQL"})
df = df.replace(r'^\s*$', np.nan, regex=True)
df
```

```python
min_bulder = df['Builder'].min()
max_bulder = df['Builder'].max()
avg_builder = df['Builder'].mean()
std_builder = df['Builder'].std()

print("Builder:")
print("Min:", min_bulder)
print("Max:", max_bulder)
print("AVG:", avg_builder)
print("AVG:", std_builder)

mim_SQL = df['SQL'].min()
max_SQL = df['SQL'].max()
avg_SQL = df['SQL'].mean()
std_SQL = df['SQL'].std()

#print(minMax_SQL)
print("SQL:")
print("Min:", mim_SQL)
print("Max:", max_SQL)
print("AVG:", avg_SQL)
print("AVG:", std_SQL)
```
